# Caso para entender el proceso de modelado y validación del ML

Cuaderno Aux:

Pasos previos -->`Entrenar`-->`Evaluar e Interpretar`-->`Decidir` --> Ir a la versión Final

Cuaderno Final v1:

1 Cargar datos-->2 Codificar-->`3 Separar`-->4 Seleccionar -->`5 Entrenar y 6 Probar`-->`7 Evaluar e 8 Interpretar`-->`9 Optimizar/10 Exportar`


## Nivel 1: Entrenar y Evaluar

```python
# Nivel 1: Junior
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
# 1) Cargar datos en X,y
personas = pd.read_csv("salario.csv") 
print(personas.head().values) # Todos los datos están limpios y codificados
target = "ingreso"
X = personas.drop(columns=[target])
y = personas[target]
print(f"target: {target}")
print(X.head().values)
print(y.head().values)

# 2) Codificar y normalizar
# 3) Separar
# 4) Seleccion de modelo(s) 
bosque = RandomForestClassifier()
# 5) Entrenar 
bosque.fit(X,y)
# 6) Probar 
# no require probar inicialmete, se hace con el mejor modelo luego de las técnicas de evaluación (en teoría)
# 7x) Evaluar sin CV, esto no es correcto
score_sin_cv=bosque.score(X,y) #hace prueba interna
print(f"score_sin_cv: {score_sin_cv}")
# 7) Evaluar
scores=cross_val_score(bosque,X,y,cv=5) #hace prueba interna
line = f"scores: {scores} | Exactitud de validación promedio (media CV): {scores.mean():.3f}"
print(line)

# 8) Interpretar
# 9) Optimizar
# 10) Exportar modelo
```

## Nivel 2: Entrenar, Probar y Evaluar 

```python
# Nivel 2: Semi-Senior

import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.metrics import confusion_matrix, classification_report

# 1) Cargar datos en X,y
personas = pd.read_csv("salario.csv")
print(personas.head().values)
target = "ingreso" # ingreso: {"Ingreso bajo": 0, "Ingreso alto": 1} 
X = personas.drop(columns=[target]) # edad: Edad de la persona, estudio: Num de años de estudio
y = personas[target]   
print(f"target: {target}")
print(X.head().values)
print(y.head().values)

# 2) Codificar y normalizar
# 3) Separar
# 4) Seleccion de modelo(s) 
bosque = RandomForestClassifier()

# 5) Entrenar con todo el conjunto, se recomienda separar 80 20
bosque.fit(X,y)

# 6x) Probar, por esta única vez de manera conveniente u optimista, es decir, predecir sobre el mismo conjunto de entrenamiento, esto no es correcto
y_pred = bosque.predict(X) # Probar con X_test

# 7x) Evaluar en el conjunto de TODO, esto no es correcto, aquí sí requiere pasar por el paso 6: Probar
reporte = classification_report(y, y_pred)
print("\nReporte de Clasificación:")
print(reporte)
print("Matriz de Confusión:")
print(confusion_matrix(y, y_pred))
print(f"\nscore: {bosque.score(X,y)}") #hace su propia prueba interna

# 7) Evaluar, como debe ser, de manera realista, mediante Validación Cruzada (CV) y con el conjunto de entrenamiento de preferencia, 
# aquí no require pasar por el paso 6: Probar
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) # Definimos la estrategia de agrupamiento
scoring = {
    "accuracy": "accuracy",
    "f1_macro": "f1_macro",
    "roc_auc": "roc_auc",
    "average_precision": "average_precision",
}

scores=cross_validate(bosque,X,y,cv=cv, scoring=scoring, return_train_score=True, n_jobs=-1) #hace prueba interna

print(f"scores: {scores['test_accuracy']} | Exactitud de validación (media CV): {scores['test_accuracy'].mean():.3f} ")
print(f"Exactitud entrenamiento (media CV): {scores['train_accuracy'].mean():.3f}")

print(f"de nuevo pero sólo promedio y con f1")
line = f"ACC {scores['test_accuracy'].mean():.3f} ± {scores['test_accuracy'].std():.3f} | F1 {scores['test_f1_macro'].mean():.3f} ± {scores['test_f1_macro'].std():.3f}"
print(line)

# Medias de las métricas ROC/PR
print(f"\n")
print(f"ROC AUC medio: {scores['test_roc_auc'].mean():.3f}")
print(f"PR Average Precision (AUPRC) medio: {scores['test_average_precision'].mean():.3f}")

# 8) Interpretar
# 9) Optimizar
# 10) Exportar modelo
```

## Nivel 3: Separar, Entrenar, Probar y Evaluar 

Nivel 3: Senior


### Parte Inicial: Entrenar

```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.metrics import confusion_matrix, classification_report

# 1) Cargar datos en X,y
personas = pd.read_csv("salario.csv")
print(personas.head().values)
target = "ingreso" # ingreso: {"Ingreso bajo": 0, "Ingreso alto": 1} 
X = personas.drop(columns=[target]) # edad: Edad de la persona, estudio: Num de años de estudio
y = personas[target]
print(f"target: {target}")
print(X.head().values)
print(y.head().values)

# 2) Codificar y normalizar
# 3) Separar
# Stratified split to preserve class proportions
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42 #, stratify=y
)
# 4) Seleccion de modelo(s) 
bosque = RandomForestClassifier()

# 5) Entrenar (con el conjunto de entrenamiento)
bosque.fit(X_train,y_train)

# 6) Probar (con el conjunto de test)
# no require inicialmete, se hace con el mejor modelo
```

### Parte Auxiliar: Evaluar y decidir

```python
# 7) Evaluar parte 2, mediante Validación Cruzada (CV) y con el conjunto de entrenamiento
# vamo con mixto
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) # Definimos la estrategia de agrupamiento
scoring = {
    "accuracy": "accuracy",
    "f1_macro": "f1_macro",
    "roc_auc": "roc_auc",
    "average_precision": "average_precision",
}

scores=cross_validate(bosque, X_train, y_train, cv=cv, scoring=scoring, return_train_score=True, n_jobs=-1) #hace prueba interna

print(f"scores: {scores['test_accuracy']} | Exactitud de validación (media CV): {scores['test_accuracy'].mean():.3f} ")
print(f"Exactitud entrenamiento (media CV): {scores['train_accuracy'].mean():.3f}")

print(f"de nuevo pero sólo promedio y con f1")
line = f"ACC {scores['test_accuracy'].mean():.3f} ± {scores['test_accuracy'].std():.3f} | F1 {scores['test_f1_macro'].mean():.3f} ± {scores['test_f1_macro'].std():.3f}"
print(line)

# Medias de las métricas ROC/PR
print(f"\n")
print(f"ROC AUC medio: {scores['test_roc_auc'].mean():.3f}")
print(f"PR Average Precision (AUPRC) medio: {scores['test_average_precision'].mean():.3f}")

# 7) Interpretar
# 8) Exportar modelo
```

### Parte 2: Probar y Evaluar

```python
# 6) Probar (con el conjunto de test)
y_pred = bosque.predict(X_test) # Probar con X_test

# 7) Evaluar en el conjunto de entrenamiento
# 7.1 Reporte y Matriz de Confusión
reporte = classification_report(y_test, y_pred)
print("\nReporte de Clasificación:")
print(reporte)
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
print(f"\nscore: {bosque.score(X_test,y_test)}") #hace su propia prueba interna
```

## 📊 Interpretación de Resultados del Modelo

### 🔹 1. Matriz de Confusión
\[
\begin{bmatrix}
22 & 1 \\
5  & 2
\end{bmatrix}
\]

- **Clase 0 (Ingreso bajo / Sin ingreso)**  
  - 22 verdaderos negativos (bien clasificados).  
  - 1 falso positivo (predijo ingreso alto pero era bajo).  

- **Clase 1 (Ingreso alto / Con ingreso)**  
  - 2 verdaderos positivos.  
  - 5 falsos negativos (no los detectó).  

👉 El modelo reconoce muy bien la clase 0, pero se le escapan muchos casos de la clase 1.

---

### 🔹 2. Métricas por clase
- **Clase 0 (mayoritaria)**  
  - Precisión = 0.81  
  - Recall = 0.96  
  - F1 = 0.88 → excelente rendimiento.  

- **Clase 1 (minoritaria)**  
  - Precisión = 0.67  
  - Recall = 0.29  
  - F1 = 0.40 → bajo rendimiento (muchos falsos negativos).  

- **Accuracy global** = 0.80 (80%).  
⚠️ El valor de accuracy es engañoso porque la mayoría de los datos son clase 0.

---

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc

# ================================
# Probabilidades de pertenecer a clase positiva (Ingreso alto = 1)
# ================================
y_prob = bosque.predict_proba(X_test)[:, 1]  # columna de probabilidad de la clase 1

# ================================
# CURVA ROC
# ================================
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"ROC curve (área = {roc_auc:.3f})")
plt.plot([0,1], [0,1], "k--", label="Azar")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("Curva ROC")
plt.legend(loc="lower right")
plt.show()

# ================================
# CURVA PRECISION-RECALL
# ================================
precision, recall, _ = precision_recall_curve(y_test, y_prob)
pr_auc = auc(recall, precision)

plt.figure(figsize=(6,5))
plt.plot(recall, precision, label=f"PR curve (área = {pr_auc:.3f})")
plt.xlabel("Recall (Sensibilidad)")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall")
plt.legend(loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc, average_precision_score

# Probabilidades para ambas clases
y_prob = bosque.predict_proba(X_test)

# =========
# ROC ambas clases
# =========
plt.figure(figsize=(6,5))
for clase in [0, 1]:
    fpr, tpr, _ = roc_curve(y_test, y_prob[:, clase], pos_label=clase)
    roc_auc = roc_auc_score((y_test == clase).astype(int), y_prob[:, clase])
    plt.plot(fpr, tpr, label=f"ROC clase {clase} (AUC={roc_auc:.3f})")

plt.plot([0,1], [0,1], "k--", label="Azar")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("Curvas ROC por clase")
plt.legend(loc="lower right")
plt.show()

# =========
# PR ambas clases
# =========
plt.figure(figsize=(6,5))
for clase in [0, 1]:
    precision, recall, _ = precision_recall_curve((y_test == clase).astype(int), y_prob[:, clase])
    pr_auc = auc(recall, precision)
    ap = average_precision_score((y_test == clase).astype(int), y_prob[:, clase])
    prevalencia = (y_test == clase).mean()

    plt.plot(recall, precision, label=f"PR clase {clase} (AP={ap:.3f} | AUC={pr_auc:.3f})")
    plt.axhline(prevalencia, linestyle="--", linewidth=1, label=f"Prevalencia clase {clase} = {prevalencia:.2%}")

plt.xlabel("Recall (Sensibilidad)")
plt.ylabel("Precision")
plt.title("Curvas Precision-Recall por clase")
plt.xlim(0, 1); plt.ylim(0, 1.05)
plt.legend(loc="upper right")
plt.show()
```

# 8) Interpretar


### 🔹 3. Curvas ROC
- **AUC clase 0 = 0.575**  
- **AUC clase 1 = 0.575**  

👉 El modelo apenas supera el azar (0.5).  
Esto indica que no separa con claridad las clases.

---

### 🔹 4. Curvas Precision-Recall
- **Clase 0 (76.7% de prevalencia)**  
  - AP = 0.790, AUC = 0.799  
  - Muy por encima de la línea base → predice bien la clase mayoritaria.  

- **Clase 1 (23.3% de prevalencia)**  
  - AP = 0.399, AUC = 0.336  
  - Apenas supera (y a veces no) la prevalencia → el modelo no identifica bien los positivos.  

---

### 🔹 5. Conclusiones
- El modelo está **sesgado hacia la clase 0 (Ingreso bajo)**.  
- El accuracy (80%) refleja sobre todo que la clase 0 es mayoría.  
- Para mejorar la clase 1 (Ingreso alto) se recomienda:  
  - Usar `stratify=y` en `train_test_split` para mantener proporción de clases.  
  - Probar `class_weight="balanced"` en el RandomForest.  
  - Usar técnicas de **balanceo de datos** (oversampling, SMOTE, undersampling).  
  - Ajustar el **umbral de decisión** para aumentar el recall de la clase 1.  
  - Probar modelos más avanzados (XGBoost, LightGBM).
